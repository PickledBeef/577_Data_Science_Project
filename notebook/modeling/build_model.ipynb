## Data Preprocessing

```python
df_notna = df.loc[df["Number of first-time undergraduates - in-state"].notna()]
df_notna.isnull().sum()
```

```python
df_notna = df_notna.loc[df["ACT Composite 25th percentile score"].notna()]
df_notna.isnull().sum()
```

```python
df_notna = df_notna.loc[df["Graduation rate - Bachelor degree within 4 years, total"].notna()]
df_notna.isnull().sum()
```

```python
df_notna = df_notna.loc[df["Percent of freshmen receiving any financial aid"].notna()]
df_notna.isnull().sum()
```
